````python
import datetime
import requests
import json
import os

from dotenv import load_dotenv
import pandas as pd
import sqlalchemy as sa
from sqlalchemy import create_engine, text, MetaData, Table
from sqlalchemy.orm import sessionmaker
````

````python
def create_connection():
    """
    Створює підключення через SQLAlchemy
    """
    # Завантажуємо змінні середовища
    load_dotenv()

    # Отримуємо параметри з environment variables
    host = os.getenv('DB_HOST', 'localhost')
    port = os.getenv('DB_PORT', '3306')
    user = os.getenv('DB_USER')
    password = os.getenv('DB_PASSWORD')
    database = os.getenv('DB_NAME')

    if not all([user, password, database]):
        raise ValueError("Не всі параметри БД задані в .env файлі!")

    # Створюємо connection string
    connection_string = f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}"

    # Створюємо engine з connection pooling
    engine = create_engine(
        connection_string,
        pool_size=2,           # Розмір пулу підключень
        max_overflow=20,        # Максимальна кількість додаткових підключень
        pool_pre_ping=True,     # Перевірка підключення перед використанням
        echo=False              # Логування SQL запитів (True для debug)
    )

    # Тестуємо підключення
    try:
        with engine.connect() as conn:
            result = conn.execute(text("SELECT 1"))
            result.fetchone()

        print("✅ Підключення до БД успішне!")
        print(f"🔗 {user}@{host}:{port}/{database}")
        print(f"⚡ Engine: {engine}")

        return engine

    except Exception as e:
        print(f"❌ Помилка підключення: {e}")
        return None

# Створюємо підключення
engine = create_connection()
````

Використайте `text()` та named parameters для формування запиту з SQLAlchemy.

Після отримання даних з БД проведіть обчислення з Python та напишіть висновки:
1. Побудуйте стовпчасту діаграму суми замовлень по країнам і напишіть, в якій країні найбільша сума замовлень за 2004 рік.
2. В країні з найбільшою кількістю замовлень знайдіть клієнта, який зробив замовлень на найбільшу суму і виведіть імʼя цього клієнта, на яку суму він зробив замовлень і який % від всіх замовлень в цій країні становить його сума замовлень за рік.

````python
from sqlalchemy import text

safe_query = text("""
            SELECT orders.orderNumber, orderDate, status, customerName, country, 
            SUM(orderdetails.quantityOrdered * orderdetails.priceEach) as total
            FROM (orders 
            LEFT JOIN customers   ON orders.customerNumber = customers.customerNumber)
            LEFT JOIN orderdetails   ON orders.orderNumber = orderdetails.orderNumber
            WHERE orderDate BETWEEN :start_date AND :end_date
            GROUP BY  orders.orderNumber
            """)

start_date = datetime.date(2004, 1, 1)
end_date = datetime.date(2004, 12, 31)
      
df_orders = pd.read_sql(
    safe_query, 
    engine,
    params={
        'start_date': start_date,
        'end_date': end_date
})

````

````python
#Побудуйте стовпчасту діаграму суми замовлень по країнам і напишіть, в якій країні найбільша сума замовлень за 2004 рік.
import pandas as pd
import matplotlib.pyplot as plt

orders_of_country = df_orders.groupby(['country'])['total'].sum()
ax= orders_of_country.plot.bar(
    figsize=(12, 6), 
    title='Orders by country for 2004',
    xlabel='Country',
    ylabel='Total',
    grid=True
)
for container in ax.containers:
    labels = [f'{val/1000:,.0f} k' for val in container.datavalues]
    ax.bar_label(container, labels=labels)
plt.show()
````

````python
print (f'Найбільша сума замовлень в {orders_of_country.idxmax()} на загальну суму {orders_of_country.max()} USD')
````

````python
#В країні з найбільшою кількістю замовлень знайдіть клієнта, який зробив замовлень на найбільшу суму і виведіть імʼя цього клієнта,
# на яку суму він зробив замовлень і який % від всіх замовлень в цій країні становить його сума замовлень за рік

clients_of_top_coutry = df_orders[df_orders['country']==orders_of_country.idxmax()].groupby('customerName')['total'].sum()

top_client =clients_of_top_coutry.idxmax()

percentage_top_client=clients_of_top_coutry.max()/orders_of_country.max()*100

print (f'Найбільша сума замовлень у клієнта {clients_of_top_coutry.idxmax()}' )
print (f'Замовлень зроблено на суму {int(clients_of_top_coutry.max())} USD' )
print (f'Частка в загальному обсязі замовлень становить {int(percentage_top_client)}%' )
````

Візуалізацію можна створювати з будь-якою бібліотекою на ваш вибір.

````python
Top_products = text("""
            WITH total_prod AS (
               SELECT 
               productName,
               productLine,
               SUM(o.quantityOrdered * o.priceEach) as total_by_product
             FROM (products p  
             LEFT JOIN orderdetails o   ON p.productCode  = o.productCode )
             GROUP BY  p.productCode
             ),
             avg_of_line AS (
                 Select *, AVG(total_by_product) Over (Partition by productLine) as avg_of_productline
                 From total_prod
             )
             Select *, 
                    RANK () Over(order by total_by_product DESC) as product_rank,
                    Round(total_by_product/SUM(total_by_product ) OVER()*100,2) as percent_of_total,
                    Round(((total_by_product/avg_of_productline)-1)*100,2) as diff_avg_of_line
             FROM avg_of_line
             ORDER BY percent_of_total DESC
              """)
df_Top_products = pd.read_sql(
    Top_products, 
    engine)
````

````python
print (f'ТОП-1 продукт це - {df_Top_products.productName[0]}, частка від загального доходу складає {df_Top_products.percent_of_total[0]} %' )
````

````python
#Створіть кругову діаграму розподілу доходу по лініях продуктів. 
#Який відсоток від всіх продажів становлять продажі за ТОП2 лініями сумарно?
lineproducts_pie = df_Top_products.groupby('productLine')['total_by_product'].sum().sort_values(ascending=False)
lineproducts_pie.plot.pie(
    figsize=(8, 8),
    autopct='%1.2f%%',
    title='Діаграма розподілу доходу по лініях продуктів',
    ylabel='',
    colors = plt.cm.Dark2.colors
)
plt.show()

TOP2_percent=((lineproducts_pie.iloc[0]+lineproducts_pie.iloc[1])/lineproducts_pie.sum())*100
print(f' TOP-2 лінії сумарно становлять {int(TOP2_percent)} % від загальної суми прдажів.')
````
